```python
import numpy as np
import pandas as pd
from PIL import Image 
import matplotlib.pylab as plt
import matplotlib.pyplot
from sklearn.decomposition import PCA

from sklearn.cluster import SpectralClustering
from sklearn.manifold import SpectralEmbedding
from sklearn.manifold import spectral_embedding

from sklearn.metrics import pairwise_distances

from sklearn.cluster import KMeans

from scipy.linalg import fractional_matrix_power
```

#### Define reusable functions: ####

```python
# opens image given file name:
def img_to_array(image_name):
    image_gray = Image.open("./data/{}".format(image_name)).convert('LA')
    reduced_image = image_gray.resize((50,50))
    reduced_image_array = np.array(reduced_image)[:,:,0]
    return reduced_image_array
```

```python
# function re-shapes the array so that it is in R3:
    # first two features are the locations of the pixels
    # third feature is the value of the pixel on the gray scale 
    # returns reshaped array in R3
def reshape_array(X):
    height, length = np.shape(X)
    i_coords, j_coords = np.meshgrid(range(height), range(length), indexing='ij')
    i_coords,j_coords
    height_array = i_coords.flatten()
    length_array = j_coords.flatten()

    gray_scale_array = X.flatten()
    X_trans = np.array([height_array,length_array,gray_scale_array])
    X_s = np.transpose(X_trans)
    
    return X_s
```

```python
# form normalized Laplacian
def form_laplacian(X,r, sigma_i, sigma_x):
    n = np.shape(X)[0]
    Y = np.zeros((n,1))
    X = np.concatenate((X,Y),axis=1)
    dist_X = pairwise_distances(X[:,0:2],metric='euclidean')
    
    # impose distance threshold
    dist_X[dist_X > r] = 'inf'
    color_dist_X = pairwise_distances(X[:,2:4],metric='euclidean')
    
    W = np.multiply(np.exp(-(color_dist_X**2)/sigma_i**2), np.exp(-(dist_X**2)/sigma_x**2))
    D = np.diag(np.sum(W,axis=0))
    L = fractional_matrix_power(D,-0.5)@(D-W)@fractional_matrix_power(D,-0.5)
    
    return W, L
```

```python
# construct weight using parameters
def construct_W(X,r, sigma_i, sigma_x):
    n = np.shape(X)[0]
    Y = np.zeros((n,1))
    X = np.concatenate((X,Y),axis=1)
    dist_X = pairwise_distances(X[:,0:2],metric='euclidean')
    
    # impose distance threshold
    dist_X[dist_X > r] = 'inf'
    color_dist_X = pairwise_distances(X[:,2:4],metric='euclidean')
    
    W = np.multiply(np.exp(-(color_dist_X**2)/sigma_i**2), np.exp(-(dist_X**2)/sigma_x**2))
    return W
```

```python
def embedding(L,c):
    
    # V is the spectral embedding of data
    e_val,e_vec = np.linalg.eig(L)
    
    # order eigenvectors in ascending order according to corresponding eigenvalue
    ind = np.argsort(e_val)
    e_val = np.real(e_val[ind])
    e_vec[:] = e_vec[:,ind]
    V = e_vec[:,0:c+1]
    V = np.transpose(V)
    V = np.real(V)
    return V,e_val
```

```python
def segment_image(X, c, k):
    leaf_clusters = KMeans(n_clusters=k,init='k-means++',).fit(X[:,0:c])
    return leaf_clusters.labels_
```

```python
def perform_segmentation(X,c,k,sigma_i,sigma_x):
    W = construct_W(X, 20, sigma_i,sigma_x)
    # perform spectral embedding
    X_embedding = spectral_embedding(W,n_components=c)
    # use k means clustering to segment image
    labels = segment_image(X_embedding,c, k)
    
    return labels
```

```python
# plot segmented image
def plot_segmented_image(labels):
    img = np.reshape(labels,(50,50))
    plt.imshow(img)
```

### Parameter Tuning: ###
- how do we measure the quality of segmentation?
- my tentative solution is to use a nromalized cut value

```python
# construct a normalized weight matrix that takes away the effect of the parameters
# not used for clustering, only used for evaluating quality of segmentation
def construct_norm_W(X,r):
    n = np.shape(X)[0]
    Y = np.zeros((n,1))
    X = np.concatenate((X,Y),axis=1)
    dist_X = pairwise_distances(X[:,0:2],metric='euclidean')
    
    # impose distance threshold
    dist_X[dist_X > r] = 'inf'
    color_dist_X = pairwise_distances(X[:,2:4],metric='euclidean')
    
    W = np.multiply(np.exp(-(color_dist_X**2)), np.exp(-(dist_X**2)))
    return W
```

```python
# function calculates the cut for a segmentation
def find_cut_val(clusters,W):
    
    # construct D
    D = np.sum(W, axis=0)
    sum_D = np.sum(D)

    # sum degrees of data points
    cluster_1_id = np.argwhere(clusters>0)
    cluster_1_id = cluster_1_id.T[0]
    sum_D_cluster_1 = np.sum(D[cluster_1_id])
    
    # construct k, b and y
    k = sum_D_cluster_1/sum_D
    b = k/(1-k)
    y = (clusters+1) - b*(1-clusters)
    
    # convert D back to diagonal matrix
    D = np.diag(D)
    
    # calculate cut
    cut = (y@(D-W)@y.T)/(y@D@y.T)
    
    return cut
    
```

```python
# function performs segmentation
def calc_cut(sigma_i, sigma_x):
    # form weight matrix
    W = construct_W(X_s, 20, sigma_i, sigma_x)
    # perform spectral embedding
    X_embedding = spectral_embedding(W,n_components=2)
    # use k means clustering to segment image
    test_segment = segment_image(X_embedding,2, 2)
    # find cut
    W_norm = construct_norm_W(X_s,20)
    cut = find_cut_val(test_segment, W_norm)
    return cut
```

```python
param_values = [2,5,10,50,100]
cut_values = np.array([])

for i in param_values:
    for x in param_values:
        cut_ix = calc_cut(i,x)
        cut_values = np.append(cut_values,[i,x,cut_ix],axis=0)
        
```

```python
cut_values = np.reshape(cut_values,(25,3))
cut_values = np.round(cut_values,5)
cut_values_df = pd.DataFrame(data=cut_values, columns = ["sigma_i","sigma_x","cut value"])
cut_values_df = cut_values_df.pivot(index="sigma_i",columns="sigma_x",values="cut value")
cut_values_df
```
